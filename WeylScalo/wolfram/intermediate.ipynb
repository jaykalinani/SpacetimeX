## Check Expression of $R_{ij}$

$$
\begin{align}
R_{jk}
&=
\partial_i\Gamma^i{}_{jk}-\partial_j\Gamma^i{}_{ki}
+\Gamma^i{}_{ip}\Gamma^p{}_{jk}-\Gamma^i{}_{jp}\Gamma^p{}_{ik}
\\
&=
g^{il}
\left[
\frac{1}{2}
\left(
-\partial_i\partial_lg_{jk}
-\partial_j\partial_kg_{il}
+\partial_j\partial_ig_{kl}
+\partial_k\partial_ig_{jl}
\right)
+\Gamma^m{}_{ji}\Gamma_{mkl}-\Gamma^m{}_{jk}\Gamma_{mil}
\right]
\end{align}
$$

```
IndexSet[EQ1[-a_, -b_], 
    RicciCD[-a, -b] - 
    metricg[c, d] (1/2 (-PD[-c][PD[-d][metricg[-a, -b]]] - PD[-a][PD[-b][metricg[-c, -d]]] + PD[-a][PD[-c][metricg[-b, -d]]] + PD[-b][PD[-c][metricg[-a, -d]]]) + 
                   ChristoffelCD[e, -a, -c] ChristoffelCD[-e, -b, -d] - ChristoffelCD[e, -a, -b] ChristoffelCD[-e, -c, -d])
];
```

## Check Expression of $\perp \mathcal{R}_{\mathbf{n}jkl}$

$$
\begin{align}
\perp\mathcal{R}_{\mathbf{n}jkl}
&=D_kK_{jl}-D_lK_{jk}
\\
&=\partial_kK_{jl}-\partial_lK_{jk}+\Gamma^m{}_{jl}K_{km}-\Gamma^m{}_{jk}K_{lm}
\end{align}
$$

```
IndexSet[EQ2[-j_, -k_, -l_],
    (PD[-k][ExtrinsicKmetrich[-j, -l]] - PD[-l][ExtrinsicKmetrich[-j, -k]] + Christoffelcd[m, -j, -l] ExtrinsicKmetrich[-k, -m] - Christoffelcd[m, -j, -k] ExtrinsicKmetrich[-l, -m]) -
    (cd[-k][ExtrinsicKmetrich[-j, -l]] - cd[-l][ExtrinsicKmetrich[-j, -k]])
];
```
